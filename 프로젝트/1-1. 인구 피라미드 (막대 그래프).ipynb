# 1-1. 인구 피라미드 (막대 그래프)

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 남자 데이터 정의

```python
file_name = '2022_연령별인구현황.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,234,567 -> 1234567 (정수형)
df_m.head(3)
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int) # 1,234,567 -> 1234567 (정수형)
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # columns 명이 같아서 .1 이 붙음
```

```python
df_w.columns = df_m.columns # columns 명 동일화
df_w.columns
```

## 데이터 시각화

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000, label='남자') # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000, label='여자')
plt.legend()
plt.title('2022년 대한민국 인구 피라미드')
plt.savefig('2022_인구피라미드(막대 그래프).png', dpi=100)
```
