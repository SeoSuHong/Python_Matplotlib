# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=100, counterclock=False)
plt.show()

# autopct : 자동으로 퍼센트 값을 소숫점 자리까지 나타냄 (%.소숫점 자리수f%%(%표시))
# startangle : 시작지점을 100도 돌림
# counterclock=False : 시계방향으로 나타냄
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C++', 'ETC']
#explode = [0.2, 0.1, 0, 0, 0, 0] # 각 데이터 별로 간격 설정
explode = [0.05] * 6 # 전체 데이터 동일한 간격 설정

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.title('언어별 선호도')
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```

```python
values = [30, 25, 20, 10, 13, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C++', 'ETC']

plt.title('언어별 선호도')
explode = [0.07] * 6
plt.pie(values, labels=labels, autopct='%.1f%%', counterclock=False, startangle=110, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```
